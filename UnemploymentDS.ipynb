# **Importing datasets**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# loading the datasets
data_1 = pd.read_csv('Unemployment_in_India[1].csv')
data_2 = pd.read_csv('Unemployment_Rate_upto_11_2020[1].csv')
data_1.columns = data_1.columns.str.strip()
data_2.columns = data_2.columns.str.strip()
```

# **Dataset 1**

```python
# EDA for Dataset 1
print("Dataset 1 Info:")
print(data_1.info())
```

```python
# checking for missing values
print(data_1.isnull().sum())
```

```python
# Visualize Dataset 1
plt.figure(figsize=(12, 6))
sns.countplot(data=data_1, x='Region', order=data_1['Region'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Region Distribution in Dataset 1")
plt.show()
```

```python
#analysing unemployment trends
grouped_data = data_1.groupby(['Region', 'Date'])['Estimated Unemployment Rate (%)'].mean().reset_index()
print(grouped_data.head())
```

```python
print(data_1.head())
print(data_1.info())
```

```python

region = 'Andhra Pradesh'  # visualizing for only andhra pradhesh region
region_data = grouped_data[grouped_data['Region'] == region]
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Estimated Unemployment Rate (%)', data=region_data)
plt.title(f'Unemployment Rate Trends in {region}')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.grid(True)
plt.show()
```
